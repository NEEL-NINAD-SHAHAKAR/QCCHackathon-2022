```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector,  array_to_latex
```

```python
#Initializing Quantum Circuit
qc = QuantumCircuit(6,6) #Defining 6 quantum register and 6 classical registers for our purpose

#Setting Input |11>

qc.x([0,1])
qc.barrier()
```

```python
#Drawing our Input Circuit
qc.draw()
```

```python
#Encoding the data

"""
As the encoding mentioned in the data sheet

The encoding required was (XZXH|q0>)(|q1>)

There is a catch, as for the state where the q0 is |1>, there should be a phase change.
Although we see its not satisfied for states with |q0> = |1> due to change in phase.

Hence we will first invert it and then teleport the same.

"""

qc.measure(0,0) #To know the status of |q0>

qc.x(0).c_if(0,1) # Phase correction

#Applying the encoding
qc.x(0)
qc.z(0)
qc.x(0)
qc.h(0)

qc.barrier()
```

```python
#We can see our Input Here

display(qc.draw())

svsim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Statevector} = "))
plot_bloch_multivector(final_state)
```

```python
#Teleporting Protocol

#Entagling qubits to be sent i.e |q2> with |q4> and |q3> with |q5>

#Entangling q2, q4
qc.h(2)
qc.cx(2,4)

#Entangling q3,q5
qc.h(3)
qc.cx(3,5)

qc.barrier()
qc.draw()
```

```python
# Bell measurement on the information to be teleported

#Measuring q0 to q2
qc.cx(0,2)
qc.h(0)

#Measuring q1 to q3
qc.cx(1,3)
qc.h(1)

qc.barrier()
qc.draw()
```

```python
#Completing the protocol by measurement to send to other end using cx and cz gate

#Teleporting |q0> to |q4>
qc.cx(2,4)
qc.cz(0,4)

#Teleporting |q1> to |q5>
qc.cx(3,5)
qc.cz(1,5)

qc.barrier()
qc.draw()
```

```python
"""
Now as the information is been succesfully teleported, now we will decrypt the information.
This is done by applying Gates in reverse order than that of encryption
"""


#Applying the decoding
qc.h(4)
qc.x(4)
qc.z(4)
qc.x(4)

qc.x(4).c_if(0,1) # Phase correction

qc.barrier()
```

```python
#Now lets measure our state and see its final state

qc.measure([4,5],[4,5])
qc.draw()

```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, backend=simulator, shots=1024).result()
plot_histogram(result.get_counts(qc))
```
